```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('F:\Python\Python Code\Pandas\student_scores.csv')
Hours=data[['Hours']]
Scores=data['Scores']
plt.scatter(Hours,Scores)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Hours,Scores,test_size=0.2,random_state=5)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
print("m=",model.coef_,"c=",model.intercept_)
plt.scatter(xtrain,ytrain,color='Red')
plt.plot(xtrain,xtrain*model.coef_+model.intercept_)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('r2_score error')
```
